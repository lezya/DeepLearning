```python
import os
import glob
import tensorflow as tf
import tensorflow.keras as k
import matplotlib.pyplot as plt
import numpy as np
import cv2
import random
```

```python
my_list = os.listdir('train')
y_data = my_list
#print(y_data)
datadir='train'
```

```python
training_data=[]

def create_training_data():
    for training in y_data:
        path=os.path.join(datadir, training)
        class_num = y_data.index(training)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (img_size, img_size))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass

create_training_data()            
```

```python
img_size = 30
new_array = cv2.resize(img_array, (img_size, img_size))
plt.imshow(new_array, cmap = 'gray')
```

```python
random.shuffle(training_data)
#class_num
```

```python
for sample in training_data:    #Rechecking the data
    print(sample[1])
```

```python
X = []
y = []
for features, label in training_data:
    X.append(features)
    y.append(label)
X = np.array(X).reshape(-1, img_size, img_size,1)
y = np.array(y)
X=X/255.0
X.shape
```

```python
model = k.Sequential()

model.add(k.layers.Conv2D(32,(3,3), padding='same', use_bias=False, input_shape=X.shape[1:]))
model.add(k.layers.LeakyReLU(alpha=0.1))
model.add(k.layers.MaxPooling2D(pool_size=(2,2)))
model.add(k.layers.Conv2D(64,(3,3), padding='same', use_bias=False))
model.add(k.layers.LeakyReLU(alpha=0.1))
model.add(k.layers.MaxPooling2D(pool_size=(2,2)))
model.add(k.layers.Flatten())
model.add(k.layers.Dense(480,activation='relu'))
model.add(k.layers.Dropout(0.25))
model.add(k.layers.Dense(240,activation='relu'))
model.add(k.layers.Dropout(0.1))
model.add(k.layers.Dense(120,activation='relu'))
#model.summary()
```

```python
N_epochs = 50
batch_size = 128
```

```python
model.compile(optimizer='adam', 
              loss='mean_squared_error',
              metrics=['mae'])
```

```python
model.fit(X,y,epochs=N_epochs,batch_size = batch_size,validation_split = 0.2)
```
